```python
# PATH FOR ALL STATES IN AGGREGATED TRANSACTIONS

path = r"C:\Users\sangeetha\Downloads\pulse\data\aggregated\transaction\country\india\state"
Agg_trans = os.listdir(path)
Agg_trans
```

```python
# INITIALIZATION

Aff_df = {'State':[],'Year' : [],'Quarter': [], 'Transaction_type':[], 'Total_transaction':[],'Transaction_amount':[]}

for i in Agg_trans:
  p1 = path + '/' +  i +'/'
  agg_t_year = os.listdir(p1)

  for j in agg_t_year:
    p2 =  p1 +j + '/'
    agg_t_quarter = os.listdir(p2)

    for k in agg_t_quarter:
      p3 = p2 + k

      data = open(p3,'r')
      d = json.load(data)

      for z in d['data']['transactionData']:

        Name = z['name']
        Count = z['paymentInstruments'][0]['count']
        Amount = z['paymentInstruments'][0]['amount']
        Aff_df['State'].append(i)
        Aff_df['Year'].append(j)
        Aff_df['Quarter'].append(int(k.strip('.json')))
        Aff_df['Transaction_type'].append(Name)
        Aff_df['Total_transaction'].append(Count)
        Aff_df['Transaction_amount'].append(Amount)

Aff_trans = pd.DataFrame(Aff_df)
Aff_trans
```

```python
# PATH FOR ALL STATES IN AGGREGATED USERS

path = r"C:\Users\sangeetha\Downloads\pulse\data\aggregated\user\country\india\state"
Agg_user = os.listdir(path)
Agg_user
```

```python
# INITIALIZATION

Aff_df2 = {'State':[],'Year' : [],'Quarter': [], 'User_brand':[], 'Count':[],'Percentage':[]}

for i in Agg_user:
  p1 = path + '/' + i +'/'
  agg_t_year = os.listdir(p1)

  for j in agg_t_year:
    p2 =  p1 +j + '/'
    agg_t_quarter = os.listdir(p2)

    for k in agg_t_quarter:
      p3 = p2 + k

      data = open(p3,'r')
      d = json.load(data)

      try:

        for z in d['data']['usersByDevice']:

            brnd = z['brand']
            cont = z['count']
            per = z['percentage']

            Aff_df2['State'].append(i)
            Aff_df2['Year'].append(j)
            Aff_df2['Quarter'].append(int(k.strip('.json')))
            Aff_df2['User_brand'].append(brnd)
            Aff_df2['Count'].append(cont)
            Aff_df2['Percentage'].append(per)

      except:
        pass

Aff_user = pd.DataFrame(Aff_df2)
Aff_user
```
